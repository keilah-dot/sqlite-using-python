```python
import matplotlib.pyplot as plt
blood_sugar_men = [114, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

type = [blood_sugar_men , blood_sugar_women]
color = ['g','r']
label = ['men', 'women']
bins = [80,100, 125, 250]
plt.xlabel('Blood sugar levels')
plt.ylabel('No of patients')
plt.hist(type, bins = bins, rwidth = 5, color = color, label = label, orientation = "vertical")
plt.title("Blood Sugar Level Chart")
plt.legend(loc='upper right')
plt.show()
```
