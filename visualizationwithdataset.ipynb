```python
missing_values=["N/a","na",np.nan]
df=pd.read_csv("country_vaccinations.csv",na_values=False)
```

```python
df.isnull().sum()
```

```python
df.isnull().any()
```

*6.Using heatmap to visualize the first 100 rows*

```python
subset=df.iloc[:5200, :]
plt.figure(figsize=(12,8))
sns.heatmap(subset.isnull(),cbar=True,cmap="viridis")
plt.show()
```

```python
df.dropna()
```

```python
subset=df.iloc[:5200, :]
plt.figure(figsize=(12,8))
sns.heatmap(subset.isnull(),cbar=True,cmap="viridis")
plt.show()
```

```python
df.fillna(0)
```

```python
df.fillna(method="bfill")
```

```python
subset=df.iloc[:5200, :]
plt.figure(figsize=(12,8))
sns.heatmap(subset.isnull(),cbar=True,cmap="RdBu")
plt.show()
```

```python
df.fillna(method="ffill")
```

```python
subset=df.iloc[:5200, :]
plt.figure(figsize=(12,8))
sns.heatmap(subset.isnull(),cbar=True,cmap="YlGnBu")
plt.show()
```
