```python
student_name=["kei", "kes", "bon", "tim", "wura", "yimix", "kuza", "shar", "drago", "sasha"]
student_marks=[50, 10, 30, 45, 49, 32, 15, 28, 10, 50]

mark_perc=[]
for x in student_marks:
  result= (x/50)*100
  mark_perc.append(result)
print(mark_perc)
```

```python
from IPython.core.formatters import PlainTextFormatter
import matplotlib.pyplot as plt
def line_chart():
  plt.plot(student_name,student_marks)
  plt.title("Student Marks Chart")
  plt.xlabel("Student Name")
  plt.ylabel("Student Marks")
  plt.show()
line_chart()
```
