```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
file = files.upload()
```

```python
df = pd.read_csv('Countries Data.csv')
df.head()
```

```python
c_52=df.loc[df['year']==1952]
c_52.head()
```

```python
c_07=df.loc[df['year']==2007]
c_07.head()
```

```python
c_merge=c_52.merge(c_07, left_on='country', right_on='country')
c_merge.head()
```

```python
c_merge.drop(['year_x','year_y'],axis=1)
```

```python
c_merge['population_growth']=c_merge['population_x']
c_merge.head()
```

```python
c_merge=c_merge.sort_values('population_growth', ascending=False)
```

```python
c_merge.head()
```

```python
c_merge=c_merge.reset_index()
```

```python
c_merge.head()
```

```python
c_merge = c_merge.drop(['index'],axis=1)
```

```python
c_merge.head()
```

```python
c_merge.info()
```

```python
plt.bar(c_merge['country'],c_merge['population_growth'],color='pink')
plt.xlabel('Country')
plt.ylabel('Population Growth')
plt.title('Population Growth of Countres')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)
plt.show()
```
