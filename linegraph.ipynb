```python
import matplotlib.pyplot as plt
x = [0,5,10,15,20,25,30]
y1 = [10,15,25,35,24,12,7]
y2 = [10,12,17,28,20,13,9]
plt.plot(x,y1, linestyle = 'dashed',marker='D',color='red',label='Vehicle1')
plt.plot(x,y2, linestyle = 'dotted',marker='o',color='blue',label='Vehicle2')
plt.title("Velocity time Graph")
plt.xlabel("Time")
plt.ylim(0,40)
plt.xlim(0,35)
plt.ylabel("Velocity")
plt.legend()
plt.show()
```

```python
import numpy as np
x=np.arange(1,11,1)
y1=(2*x)+1
y2=(2*x*x)+2
plt.plot(x,y1, linestyle = 'dashed',marker='D',color='pink',label='y=2x+1')
plt.plot(x,y2, linestyle = 'dotted',marker='D',color='red',label='y=(2x^2+2')
plt.legend()
plt.show()
```
