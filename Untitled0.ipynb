```python
import matplotlib.pyplot as plt
level = [0,2,4,6,8,10,12]
firstset = [0,3,6,12,2,1,7]
secondset = [0,5,10,12,11,6,1]
plt.plot(level,firstset, linestyle = 'dashed',marker='D',color='pink',label='First class lvl')
plt.plot(level,secondset, linestyle = 'dotted',marker='D',color='purple',label='Second class lvl')
plt.title("diffrence lvl graph")
plt.xlabel("Level")
plt.ylim(0,14)
plt.xlim(0,14)
plt.ylabel("Players")
plt.legend()
plt.show()
```
