```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'Monday':12, 'Tuesday':15, 'Wednesday':11, 'Thursday':9, 'Friday':1, 'Saturday':9, 'Sunday':21}
days = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))

plt.bar(days, values, color = 'purple', width = 0.4, edgecolor='navy')

plt.xlabel("Days of births")
plt.ylabel("Number of births")
plt.title("Birth rate date of July")
plt.show()
```

```python
data = {'Monday':17, 'Tuesday':5, 'Wednesday':2, 'Thursday':11, 'Friday':1, 'Saturday':8, 'Sunday':29}
days = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))

plt.bar(days, values, color = 'navy', width = 0.4, edgecolor='purple')

plt.xlabel("Days of births")
plt.ylabel("Number of births")
plt.title("Birth rate date of August")
plt.show()
```

```python
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

July = [12, 15, 11, 9, 1, 9, 21]
August = [17, 5, 2, 11, 1, 8, 29]

br1 = np.arange(len(July))
br2 = [x + barWidth for x in br1]

plt.bar(br1, July, color = 'purple', width = barWidth, edgecolor = 'navy', label = 'July')
plt.bar(br2, August, color = 'navy', width = barWidth, edgecolor = 'purple', label = 'August')

plt.xlabel('Months', fontweight = 'bold', fontsize = 15)
plt.ylabel('Number of Births', fontweight = 'bold', fontsize = 15)
plt.xticks([r + barWidth/2 for r in range(len(July))],
           ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun'])

plt.legend()
plt.show()
```
