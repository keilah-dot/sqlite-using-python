```python
#Numpy = is for maths and arrays. Its like a calculator.
import numpy as np

scores = np.array([56, 78, 45, 90, 67, 88])

print("Scores:", scores)

print("Average:", np.mean(scores))

print("Highest:", np.max(scores))
print("Lowest:", np.min(scores))

bonus_scores = scores + 5
print("bonus:", bonus_scores)
```

```python
#Pandas = organizes messy data. Like to make tables and filter data.
import pandas as pd

data = {
    "Name": ["Alex", "Sam", "Jordan", "Taylor"],
    "Hours Studied": [2, 5, 1, 4],
    "Score": [50, 85, 40, 75]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Matplotlib = is used to plot graphs.
import matplotlib.pyplot as plt

hours = [1, 2, 3, 4, 5]
scores = [45, 50, 80, 70, 90]

plt.plot(hours, scores, linestyle ='dashed', color = 'y')
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.title("Study Time vs Score")
plt.show()
```
